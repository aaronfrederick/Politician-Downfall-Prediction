```python
# Connecting to Database
db = MongoClient('localhost', 27017)['politician_db']
collection_names = db.list_collection_names()
```

```python
# Loading Models
w2v = gensim.models.Word2Vec.load('models/word2vec.model')
km = pickle.load(open('models/kmeans.pkl', 'rb'))
```

```python
clus_one = []
clus_sixteen = []

for name in collection_names:
    for doc in db[name].find():
        try:
            text = doc['full_text']
        except:
            #print('no full text for this doc')
            continue
        vec = np.zeros(150)
        length = 0
        for word in text.split():
            try:
                vec += w2v[word]
                length += 1
            except:
                pass
        cluster = km.predict([vec/length])
        #print(cluster)
        if cluster == 1:
            clus_one.append(text)
        if cluster == 16:
            clus_sixteen.append(text)
```

```python
print(len(clus_one))
print(len(clus_sixteen))
```

```python
def create_tokens(list_of_docs):
    ret_list = []
    for doc in list_of_docs:
        for token in doc.split():
            ret_list.append(token)
    return ret_list
```

```python
one_tokens = create_tokens(clus_one)
sixteen_tokens = create_tokens(clus_sixteen)
```

```python
one_counts = pd.Series(one_tokens).map(lambda x: (x,1))
```

```python
count_dict = {}
for token in one_counts:
    try:
        count_dict[token[0]] += 1
    except:
        count_dict[token[0]] = 1
```

```python
for w in sorted(count_dict, key=count_dict.get, reverse=True):
    print(w, count_dict[w])
```

```python
six_counts = pd.Series(sixteen_tokens).map(lambda x: (x,1))
count_dict = {}
for token in six_counts:
    try:
        count_dict[token[0]] += 1
    except:
        count_dict[token[0]] = 1
for w in sorted(count_dict, key=count_dict.get, reverse=True):
    print(w, count_dict[w])
```
